```python
from pandas import read_csv
from pandas import cut
from pandas import to_numeric
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from matplotlib import pyplot as plt

filename = "E:\\Work\\FinalProjectData\\third.csv"
df  = read_csv(filename)

df.pop('name')
df.pop('address')
df.pop('phoneNumber')

le = LabelEncoder()
for col in df.columns:
    if col == 'city':
        le.fit(df[col])
        df[col] = le.transform(df[col])

df
```

```python
X = df.loc[:, ['age', 'data', 'sms', 'activeCalls', 'calls', 'city']].values
y = df.loc[:, ['plan']].values

normalized_X = preprocessing.normalize(X)
normalized_X
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=7, random_state=0).fit(normalized_X)
kmeans.labels_
kmeans.cluster_centers_
```

```python
y_kmeans = kmeans.predict(X)

plt.scatter(normalized_X[:, 1], normalized_X[:, 2], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 1], centers[:, 2], c='black', s=200, alpha=0.5)
```
